```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('train.csv')
```

```python
_corr = train_data.corr()
selected_feature = _corr.index[abs(_corr['SalePrice']) > 0.4] 
_train = train_data[selected_feature]
target = _train['SalePrice']
train_df = _train.drop(['SalePrice'], axis =1)
```

```python
train_df.fillna(train_df.mean())
```

```python
# 정규화   
normal_df  = (train_df - train_df.mean())/train_df.std() 
normal_df
```

```python
cate_train = train_data.select_dtypes(include='object')
```

```python
cate_train
```

```python
# null 일정 이상 
remove_col = ['Alley' , 'FireplaceQu', 'PoolQC' ,'Fence' ,'MiscFeature']
cate_train = cate_train.drop(remove_col, axis = 1) 
```

```python
cate_train_encode = pd.get_dummies(cate_train)
```

```python
cate_train_encode.fillna(cate_train_encode.mean())
```

```python
boston_df = pd.merge(cate_train_encode, normal_df, left_index=True, right_index=True)
```

```python
boston_df
```

```python
import lightgbm as lgb 
from sklearn.model_selection import train_test_split 

x_train, x_valid, y_train, y_valid = train_test_split(normal_df, target, test_size = 0.01, shuffle = True, random_state=34)
train_ds = lgb.Dataset(x_train, label = y_train)
valid_ds = lgb.Dataset(x_valid, label = y_valid)
```

```python
params = {'learning_rate': 0.001,
          'max_depth': -1,
          'boosting': 'dart',
          'objective': 'regression',
          'metric': 'mse',
          'is_training_metric': True,
          'num_leaves': 1024,
          'feature_fraction': 0.9,
          'bagging_fraction': 0.7,
          'bagging_freq': 5,
          'seed':2020} 

model = lgb.train(params, train_ds, 5000, valid_ds, verbose_eval=100, early_stopping_rounds=100)
```

```python
test_data = pd.read_csv('test.csv')
```

```python
selected_feature_test = selected_feature.drop('SalePrice')
_test = test_data[selected_feature_test] 
_test.fillna(_test.mean()) 
normal_test_df  = (_test - _test.mean())/_test.std()  
cate_test = test_data.select_dtypes(include='object')
cate_test = cate_test.drop(remove_col, axis = 1)  
cate_test_encode = pd.get_dummies(cate_test) 
cate_test_encode.fillna(cate_test_encode.mean()) 
test_df = pd.merge(cate_test_encode, normal_test_df, left_index=True, right_index=True)
```

```python
sub = pd.read_csv('sample_submission.csv')
```

```python
sub['SalePrice'] = model.predict(normal_test_df)
```
